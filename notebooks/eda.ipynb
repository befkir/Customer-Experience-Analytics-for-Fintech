```python
import sys
sys.path.append('../src')
sys.path.append('../utils')

from data_preprocessing import combine_reviews, inspect_reviews,clean_Export
from logger import get_logger
from visualization import plot_rating_comparison,plot_all_rating

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

logger = get_logger(__name__)
```

```python
# Load Data
df_cbe = pd.read_csv('../data/Commercial_20250607_161807.csv')
df_boa = pd.read_csv('../data/Abysinia_20250607_161803.csv')
df_dashen = pd.read_csv('../data/Dashen_20250607_161801.csv')
print("CBE:", len(df_cbe))
print("BOA:", len(df_boa))
print("Dashen:", len(df_dashen))
```

```python
# Abyssinia
logger.info("Starting EDA analysis...")
inspect_reviews(df_boa)
df_boa_cleaned = clean_Export(df_boa,'Abyssinia')
```

```python
# CBE
inspect_reviews(df_cbe)
df_cbe_cleaned = clean_Export(df_cbe,'Commercial')
```

```python
# Dashen
inspect_reviews(df_dashen)
df_dashen_cleaned = clean_Export(df_dashen,'Dashen')
```

```python
# Step 1: Count ratings for each DataFrame
boa_ratings = df_boa_cleaned['rating'].value_counts().sort_index()
commercial_ratings = df_cbe_cleaned['rating'].value_counts().sort_index()
dashen_ratings = df_dashen_cleaned['rating'].value_counts().sort_index()

# Step 2: Create a combined DataFrame for plotting
rating_df = pd.DataFrame({
    'BOA': boa_ratings,
    'Commercial': commercial_ratings,
    'Dashen': dashen_ratings
}).fillna(0)

plot_rating_comparison(rating_df)
```

```python
combined = pd.concat([df_boa_cleaned, df_cbe_cleaned, df_dashen_cleaned], ignore_index=True)

# Call function
plot_all_rating(combined)
```
